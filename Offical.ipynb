```python
import random
import json
x = open('WWTBAM_Questions.json') #opens JSON file
question_bank = json.load(x) # loads JSON file as a dictionary

# cash={1:100,2:200,3:300,4:500,5:1000,6:2000,7:4000,8:8000,9:16000,10:32000,11:64000,12:125000,13:250000,14:500000,15:1000000}
class WWTBAM():
    
    def __init__(self):
        self.question=0 # total amount of questions answered 
        self.cash={0:0,1:100,2:200,3:300,4:500,5:1000,6:2000,7:4000,8:8000,9:16000,10:32000,11:64000,12:125000,13:250000,14:500000,15:1000000} # cash values
        self.correct=0 # total amount of questions correct 
        self.wrong=0 # total amount of question wrong
    
    def how_to_play(self): # directions for how to play
        print("\nTo win the million dollars, you will need to answer 15 questions correctly in a row.")
        print("After completeing 5 questions are are guaranteed $1,000, after 10 it is $32,000 and a $1,000,000 after all 15.")
        print("Before each question, you will have the option to either continue or quit and leave with a set amount of cash for the number of questions you answered correctly.")
        print("To answer a question, answer with just the letter of the correct answer. Enjoy playing!\n")
    
    def questions(self):
        global i # makes i global so it can be accessed in the other functions
        global c # makes c global so it can be accessed in the other functions
        i=1
        while i<16: # while loop that asks the user up to 15 questions
            k=i
            print(f"Would you like to continue or chose to quit and win {self.cash[k-1]}")
            c=input("If you want to continue enter 'y', if you want to quit enter 'n': ") # lets them chose to continue playing or lets them quit and leave with an amount of money
            if c=='y': # if they want to continue it asks the user a random question from the question bank
                self.question+=1
                random_question = random.choice(question_bank)

                # Print the randomly selected question
                print(f"\nQuestion {i}: {random_question['question']}")
                print("A:", random_question["A"])
                print("_______________________________________________")
                print("B:", random_question["B"])
                print("_______________________________________________")
                print("C:", random_question["C"])
                print("_______________________________________________")
                print("D:", random_question["D"])
                print("_______________________________________________")
                answer=input("What is your answer to the question? ")
                if answer == random_question["answer"] or answer.upper()==random_question["answer"]: # checks to see if the user entered the correct answer
                    print("\nCorrect!")
                    self.correct+=1
                    i+=1

                else:
                    print(f"\nWrong, you lost, the correct answer is {random_question['answer']}.") # if its wrong, it tells them what the correct answer is 
                    self.wrong+=1
                    break # breaks the while loop if they miss a question
            elif c=='n':
                print("Thank you for playing.")
                break # breaks the while loop if they decide to quit
                
    def win(self): # lets the user see their win
        if c=='y': 
            if i>5 and i<10: # if user gets between 5 and 10 questions correct but ends up missing one, they are guaranteed $1,000
                print(f"\nYou won $1,000.\n")
            elif i>10 and i<15: # if user gets between 10 and 15 questions correct but misses one, they are guaranteed $32,000
                print(f"\nYou won $32,000.\n")
            elif i>15: #if user gets 15 questions correct then they win a million dollars
                print(f"\nCongratulations, you won $1,000,000!!\n")
            else:
                print("\nYou did't win anything, good luck next time.\n")   
        else: # if user says they don't want to continue then they'll win that amount of cash for how many questions they answered correctly
            print(f"\nYou won ${self.cash[i-1]}.\n")
    
    def stats(self): # prints the players stats of how many questions they answered and how many they got correct. 
        print(f"\nYou answered a total of {self.question} questions, you got {self.correct} correct and get {self.wrong} wrong.\n")
```
